## REGRESSION FOR HOSPITAL INSURANCE DATA USING BOOSTING

```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor,VotingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split, cross_val_score, cross_validate
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
```

```python
Insurance_original = pd.read_csv(r"C:\Users\Abuja Office\Downloads\Health_Insurance_Cleaned.csv")
Insurance_tree = Insurance_original.copy()
Insurance_tree.head(3)
```

```python
# Splitting data into features matrix(X) and Target Vector(y)
x = Insurance_tree.drop(columns = ["Charges"], axis = 'columns')
y = Insurance_tree["Charges"]
```

```python
# Creating an instance of all the boosting algorithms
adab = AdaBoostRegressor(n_estimators = 500, learning_rate = 0.05, random_state = 42)
xgb = XGBRegressor(n_estimators = 500, learning_rate = 0.05, random_state = 42)
lgbm = LGBMRegressor(n_estimators = 500, learning_rate = 0.05, random_state = 42)
cat = CatBoostRegressor(n_estimators = 500, learning_rate = 0.05, random_state = 42, verbose = 0)
```

```python
# Defining the number of folds for cross validation
kfold = KFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
# Performing k-fold cross validatin using cross_validate function
adab_scores_cv = cross_validate(adab, x,y, cv = kfold, scoring = ["r2","neg_mean_absolute_error","neg_mean_squared_error","neg_root_mean_squared_error"],return_estimator = True)
xgb_scores_cv = cross_validate(xgb, x,y, cv = kfold, scoring = ["r2","neg_mean_absolute_error","neg_mean_squared_error","neg_root_mean_squared_error"],return_estimator = True)
lgbm_scores_cv = cross_validate(lgbm, x,y, cv = kfold, scoring = ["r2","neg_mean_absolute_error","neg_mean_squared_error","neg_root_mean_squared_error"],return_estimator = True)
cat_scores_cv = cross_validate(cat, x,y, cv = kfold, scoring = ["r2","neg_mean_absolute_error","neg_mean_squared_error","neg_root_mean_squared_error"],return_estimator = True)
```

```python
# ADABOOST
print("K-Fold Cross-Validation Results:")
for metric_name, scores in adab_scores_cv.items():
    if metric_name.startswith('test'):
        print(f'Metric : {metric_name.replace("test_","").capitalize()}')
        print(f'Mean : {scores.mean() * 100:.2f}%')
        print(f'Standard Deviation: {scores.std():.2f}\n')
```

```python
# XgBOOST
print("K-Fold Cross-Validation Results:")
for metric_name, scores in xgb_scores_cv.items():
    if metric_name.startswith('test'):
        print(f'Metric : {metric_name.replace("test_","").capitalize()}')
        print(f'Mean : {scores.mean() * 100:.2f}%')
        print(f'Standard Deviation: {scores.std():.2f}\n')
```

```python
# LGBM
print("K-Fold Cross-Validation Results:")
for metric_name, scores in lgbm_scores_cv.items():
    if metric_name.startswith('test'):
        print(f'Metric : {metric_name.replace("test_","").capitalize()}')
        print(f'Mean : {scores.mean() * 100:.2f}%')
        print(f'Standard Deviation: {scores.std():.2f}\n')
```

```python
# Catboost
print("K-Fold Cross-Validation Results:")
for metric_name, scores in cat_scores_cv.items():
    if metric_name.startswith('test'):
        print(f'Metric : {metric_name.replace("test_","").capitalize()}')
        print(f'Mean : {scores.mean() * 100:.2f}%')
        print(f'Standard Deviation: {scores.std():.2f}\n')
```

MODEL BUILDING(VOTING REGRESSOR)

```python
# Extracting the list of fitted models from each fold for each of the algorithm
adab_estimators = adab_scores_cv['estimator']
xgb_estimators = xgb_scores_cv['estimator']
lgbm_estimators = lgbm_scores_cv['estimator']
cat_estimators = cat_scores_cv['estimator']
```

```python
# Creating a list of tuple pairs ("model name", model)
adab_estimators_list = [(f'AdaBoost_{i}', model) for i,model in enumerate(adab_estimators)]
xgb_estimators_list = [(f'XGBoost_{i}', model) for i,model in enumerate(xgb_estimators)]
lgbm_estimators_list = [(f'LightGBM_{i}', model) for i,model in enumerate(lgbm_estimators)]
cat_estimators_list = [(f'CatBoost_{i}', model) for i,model in enumerate(cat_estimators)]
```

```python
# Building the voting for each estimator
adab_ensemble = VotingRegressor(estimators = adab_estimators_list)
xgb_ensemble = VotingRegressor(estimators =xgb_estimators_list,)
lgbm_ensemble = VotingRegressor(estimators =lgbm_estimators_list)
cat_ensemble = VotingRegressor(estimators = cat_estimators_list)
```

TRANING AND PREDICTION

```python
# Splitting data in training and validation set using train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 42)
```

```python
# Training
adab_ensemble.fit(X_train, y_train)
xgb_ensemble.fit(X_train, y_train)
lgbm_ensemble.fit(X_train, y_train)
cat_ensemble.fit(X_train, y_train)

# Prediction
y_pred_adab = adab_ensemble.predict(X_test)
y_pred_xgb = xgb_ensemble.predict(X_test)
y_pred_lgbm = lgbm_ensemble.predict(X_test)
y_pred_cat = cat_ensemble.predict(X_test)
```

```python
# Checking the performance of the model
# adab_boost
print(f"Mean Absolute Error: {round(mean_absolute_error(y_test, y_pred_adab),2)}")
print(f"Mean Squared Error : {round(mean_squared_error(y_test, y_pred_adab),2)}")
print(f"Root Mean Squared Error: {round(np.sqrt(mean_squared_error(y_test, y_pred_adab)),2)}")
print(f"Coeffiecient of determination (R2-Score): {round(r2_score(y_test, y_pred_adab),2)}")
print(f"Mean Absolute Percentage Error (MAPE): {round(np.mean(np.abs((y_test - y_pred_adab) / y_test)),2) * 100}%")
print(f"Model Accuracy: {round(adab_ensemble.score(X_test, y_test),2) * 100}%")
```
