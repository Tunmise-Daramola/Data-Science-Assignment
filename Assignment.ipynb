```python
#Assignment 7
num_1 = float(input("Enter first number"))
num_2 = float(input("Enter second number"))
operator=input("Enter operator (-,+,/,//,%):")
if operator == "+":
    print("Ans:",num_1 + num_2)
elif operator == "-":
    print("Ans:",num_1 - num_2)
elif operator == "/":
    if num_2 != 0:
        print("Ans:",num_1/num_2)
    else: print ("Error")
elif operator == "//":
    if num_2 != 0:
        print("Ans:",num_1//num_2)
    else:print("Error")
elif operator == "%":
    if num_2 != 0:
        print("Ans:",num_1 % num_2)
    else: print("Error")
```
