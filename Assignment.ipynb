```python
#1 Write a python program using a while loop that asks users bfor a number and calculates its factorial
def factorial(num):
    if num==1:
        return 1
    else:
        return num*factorial(num-1)
number = int(input("Enter a number:"))
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
#3 Write a python program taht prints all even numbers from one to 20 but skips printing number 10
value = 0 
while value<=20:
    value+= 1
    if value %2 != 0:
        continue
    elif value == 10:
        continue
    print(value)
```

```python
#4 Write a python program using for loop that takes a number as input and print whether it is a prime number or not
num = int(input("Enter a number"))
if num <= 1:
    print(num,"is not a prime number")
else:
    for i in range(2,num):
        if num % i == 0:
            print(num,"is not a prime number")
            break
    else:
        print(num,"is a prime number")
```

```python
#5 Write  a program that checks for the largest number in a list,using for loop
list = [2,27,99]
largest = list[0]
for i in list:
    if i > largest:
        largest = i
            

print("The largest number in the list is;",largest)
```

```python
#7  Write a program to sort list of tuples by second element
def fibonacci(n):
    if n == 0:
        return 0
    elif n ==1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

num = int(input("Enter a number:"))
result = fibonacci(num)
print("Fibonacci:",result)

```

- Complier: converts the whole program into machine code before running
- Interpreter: Executes code line by line without creating a separate machine code file.
- Machine language : Binary code(0s and 1s) the computer understands directly.
- Low-level language:Close to machine code;hard for humans but control hardware precisely
- High-level language: Easy-to-read language like python or java;needs a complier or interpreter.
- Assembly language:Uses symbols(mnemonics) insted of binary; converted to machine code by an assembler
